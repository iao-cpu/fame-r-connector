```r
# This program connects the fame database and retrieves data in data frame.
# It plots day wise line graph of given series for year.

library(FameRconnector)
require(ggplot2)
library(forecast)

data.df <- FAMEData('*$get_list{famedate, fis.close}', "2015", "2015", 0, "BUSINESS", "down", "no heading", "normal")

data.df <- na.omit(data.df)

plot <- ggplot(data.df, aes(x = FAMEDATE., y = as.numeric(FIS.CLOSE))) +
  geom_line() +
  labs(x = "Month(s)",
       y = "close price",
       title = "FIS.CLOSE Day wise")
print(plot)
```
