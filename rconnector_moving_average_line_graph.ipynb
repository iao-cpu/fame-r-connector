```r
# This program connects the fame database and retrieves data in data frame.
# It convers data into time series and the calculates moving average on it
# Then it plots the result againest line graph

library(FameRconnector)
require(ggplot2)
library(forecast)


FAMESetProperty("chli_log_path", "generic_connector.log")
FAMESetProperty("chli_log_enable", "True")

data.df <- FAMEData('*$get_list{fis.close}', "2002", "2015", 10, "BUSINESS", "down", "no heading", "normal")

data.df <- na.omit(data.df)

# create time series
ts_2015_2018<-ts(data=data.df$FIS.CLOSE , frequency = 12 , start = c(2002,1),end = c(2015,12))

# calculate moving average
sm <- ma(ts_2015_2018, order = 12) # 12 month moving average

# plot moving average
plot.ts(ts_2015_2018, xlab="Monthly Moving Average", ylab = "Close Price", main = "FIS.Close Price MA")

# plot
lines(sm, col="red")
```
