```r
# This program demonstrates how to create time series in 
# fame database.
# It also shows how to insert new values in time series or 
# update existing values.

library(FameRconnector)
options(digits = 12)

## open database
data <-
  FAMEData(
    "* open <access overwrite> samples_write_db.db as sample_wr --shared access",
    "2015",
    "2015",
    0,
    "monthly",
    "down",
    "heading",
    "normal"
  )
print(data)

## create time series
data <-
  createTimeSeries(
    "sample_wr",
    "sample_PRECISION_time_series",
    "PRECISION",
    "MONTHLY",
    "BUSINESS",
    "END",
    "test desc of scalar for sample"
  )
print(data)

## check if time series has got created
data <-
  FAMEData(
    "wildlist(sample_wr, \"sample_PRECISION_time_series\")",
    "1",
    "",
    1,
    "CASE",
    "down",
    "heading",
    "normal"
  )
print(data)

## check type of time series
data <-
  FAMEData("type(sample_PRECISION_time_series)",
           "1",
           "",
           1,
           "CASE",
           "down",
           "heading",
           "normal")
print(data)

dat <- c(112.11, NA, 119.56, NA, 101, 109.89, 111, 111.1, 111.56, NA, 102.65, NA)

## update time series
data <-
  updateTimeSeries("sample_wr",
                   "sample_PRECISION_time_series",
                   "JAN2022",
                   "MAR2022",
                   dat)
print(data)

## get data from time series
data <-
  FAMEData(
    "sample_PRECISION_time_series",
    "JAN2022",
    "",
    12,
    "MONTHLY",
    "down",
    "heading",
    "normal"
  )
print(data)

## get database file handle
dbfile <-
  FAMEData("filespec(sample_wr)",
           "*",
           "*",
           0,
           "case",
           "down",
           "heading",
           "normal")
dbfile <- dbfile$`filespec(sample_wr)`

## close database
data <-
  FAMEData(
    "* close sample_wr --shared access",
    "2015",
    "2015",
    0,
    "monthly",
    "down",
    "heading",
    "normal"
  )

## remove database file
file.remove(dbfile)
```
