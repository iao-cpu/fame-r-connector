```r
################################################################################
# This program demonstrates how to use fame time scaling feature
# It demonstrates conversion of data  using various techniques
# to different frequencies
################################################################################
library(FameRconnector)

print("Convert yearly data to quarterly data")
s1 <- FAMEData("*$get_list{famedate, C16676410X01.UOPEN}", "2015", "2019", 5, "QUARTERLY", "down", "heading", "normal")
print(s1)

# calculates moving average of series with business frequency and converts to annual frequency
print("")
print("Calculate moving average; convert from business frequency to annual frequency")
s1 <- FAMEData("*$get_exprs \"\"famedate, MAVE(ibm.close,30)\"\" ", "2015", "2019", 5, "Annual", "down", "heading",
              "normal")
print(s1)

# bulk retrieval; converts data from business frequency to quarterly frequency
print("")
print("Bulk retrieval; convert from business frequency to quarterly frequency")
s1 <- FAMEData('*$get_list{"famedate", fis.open, fis.low, fis.high, fis.close, ibm.open, ibm.low, ibm.high, ibm.close}', "2015", "2019", 5, "Quarterly", "down", "heading", "normal")
print(s1)

# use of convert function
print("")
print("Convert function; convert from business frequency to annual frequency")
s1 <- FAMEData("*$get_exprs \"\"famedate, convert(fis.open, ANNUAL, constant)\"\" ", "2015", "2019", 5, "Annual", "down", "heading", "normal")
print(s1)

# use of constant conversion technique along with automatic conversion off
print("")
print("Convert function with automatic conversion off")
s1 <- FAMEData("$eval_opt(\"\"convert(ibm.close, monthly, *, high)\"\" , \"\"convert automatic off\"\")", "2015", "2019", 5, "monthly", "down",
              "heading", "normal")
print(s1)
```
